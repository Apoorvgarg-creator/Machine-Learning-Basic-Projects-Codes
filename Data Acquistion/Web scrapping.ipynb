```python
import bs4
import requests
```

```python
url = 'https://www.passiton.com/inspirational-quotes?page=2'
```

```python
response = requests.get(url)
```

```python
response.url
```

```python
# response.content
```

```python
# HTML parser
soup = bs4.BeautifulSoup(response.content,'html.parser')
```

```python
# soup
```

```python
article_element = soup.find_all('a')
```

```python
article = article_element[23]
```

```python
article_element[0].img.attrs['src']
```

```python
with open('inspiration.jpg','wb') as file:
    img_url = article.img.attrs['src']
    response = requests.get(img_url)

    file.write(response.content)
```

```python
for i, article in enumerate(article_element[7:]):
    with open('inspiration{}.jpg'.format(i),'wb') as file:
        img_url = article.img.attrs['src']
        response = requests.get(img_url)

        file.write(response.content)
```
