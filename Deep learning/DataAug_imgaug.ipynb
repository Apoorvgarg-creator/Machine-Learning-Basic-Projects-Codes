## Data Augmentation using imgaug


```python
from imgaug import augmenters as iaa
```

```python
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img1 = image.load_img("African_Bush_Elephant.jpg")
img2 = image.load_img("Angel.jpg")
```

```python
img1 = image.img_to_array(img1)/255.0
img2 = image.img_to_array(img2)/255.0
```

```python
plt.imshow(img1)
plt.axis('off')
plt.show()
plt.imshow(img2)
plt.axis('off')
plt.show()
```

```python
seq = iaa.Sequential([
    iaa.Crop(px=(0,16)), # crop images from each side by 0 to 16px
    iaa.Fliplr(0.5), # horizontally flip 50% of the images
    iaa.GaussianBlur(sigma=(0,3.0)), #blur images with a sigma of 0 to 3.0
    iaa.LinearContrast((0.75,1.5)),
    iaa.Affine(
        scale = { "x": (1, 1.2), "y":(0.8,1.2)}
,       translate_percent={"x":(-0.2,0.2),"y":(-0.2,0.2)},
        rotate=(-25,25),
        shear=(-8,8)
    )
])
```

```python
batch = img1.reshape((1,*img1.shape))
batch.shape
```

```python
for i in range(6):
    output = seq(images=batch)
    # output.shape
    plt.imshow(output[0])
    plt.axis('off')
    plt.show()
```
