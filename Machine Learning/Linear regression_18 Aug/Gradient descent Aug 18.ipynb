# Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X =np.arange(10)
Y = (X-5)**2

# grad is decided by differentiating y w.r.t x
```

##### GOAL --> Given a function f(x), we want to find the value of x that minimizes f

### Visualization

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(x)")

# plt.show()
```

```python
x = 0
# step size
lr = 0.1  # lr = learning rate
error = []
# 50 steps in the downhill direction
for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad   # Formula
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print("grad -->",grad,end=" and ")
    print("x -->",x)
```

```python
# Plot the values of error

plt.plot(error)
```
