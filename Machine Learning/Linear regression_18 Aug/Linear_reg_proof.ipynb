```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0,40,2)
theta = np.array([2,3])
```

```python
noise = np.random.randn(20)*5
```

```python
y_ideal = theta[0] + theta[1]*X
y_actual = theta[0] +theta[1]*X +noise
```

```python
plt.scatter(X,y_actual,color="red",label="Actual Value")
plt.plot(X,y_ideal,color="blue",label="Predicted Value")
plt.legend()
plt.title("Linear Regression")
plt.show()
```
