```python
import cv2
import matplotlib.pyplot as plt
```

```python
img =cv2.imread('/Users/apoorvgarg/Downloads/Naruto.jpg')
```

```python
print(img.shape)
```

```python
plt.imshow(img)
```

```python
rgb_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python

plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

```python
#crop the certain part of image
plt.imshow(rgb_image)
plt.show()
```

```python
crop_image = rgb_image[0:1000,500:1500,:]
plt.imshow(crop_image)
```

```python
crop_image.shape
```

```python
total_pixels=crop_image.shape[0]*crop_image.shape[1]
print(total_pixels)
```

```python
crop_imag= cv2.resize(crop_image,(100,100))
plt.imshow(crop_imag)
plt.axis("off")
plt.title("Naruto")
plt.show()
```
