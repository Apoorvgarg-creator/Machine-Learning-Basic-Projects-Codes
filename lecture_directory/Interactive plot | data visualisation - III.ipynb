#### Interactive plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
X = pd.read_csv("/Users/apoorvgarg/PycharmProjects/CB_lecture/Challenge/Training Data_Hard/Linear_X_Train.csv")
Y = pd.read_csv("/Users/apoorvgarg/PycharmProjects/CB_lecture/Challenge/Training Data_Hard/Linear_Y_Train.csv")
```

```python
theta = np.load("/Users/apoorvgarg/PycharmProjects/CB_lecture/Challenge/Thetalist.npy")
```

```python
plt.ion()
T0 = theta[:,0]
T1 = theta[:,1]
for i in range(0,50,3):
    y_ = T1[i]*X.values +T0
    plt.scatter(X,Y.values)

    plt.plot(X,y_,"red")
    plt.draw()
    plt.pause(1)
    plt.clf()  # Destroy the last object,with the help of this we will get update graph
```
