### Seaborn
- Statistical data visualization
- sync better with pandas

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)  # It shows the mean of total bill for each sex
```

```python
sns.stripplot(x='sex',y="total_bill",data=tips)
```

```python
# change estimator
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# for count plot , we do not give y axis since it will always be the count
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# Distribution Plot -->  y axes shows frequency
# kde --> kernel density estimation
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
# to see the relationship between two variables (bipartite relation)
sns.jointplot(x='total_bill',y='tip',data=tips,kind="hex",color="red")
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="reg",color="green")
```

```python
# to compare all the features
sns.pairplot(tips)
```

```python
# hue is a very powerful and helpful option --> as you can see the data on each element and make observations
sns.pairplot(tips,hue='day')
```

```python
# VERY IMPORTANT
tips.head()
```

```python
# Creates bipartite relation --> Create correlations among the variables
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
# correlation helps in creating heat map
sns.heatmap(tips_corr,annot=True)
```

```python
sns.boxenplot(y='total_bill',x='sex',data=tips)
```
